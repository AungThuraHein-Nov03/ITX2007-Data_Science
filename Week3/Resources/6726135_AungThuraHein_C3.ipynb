```python
superstore = pd.read_csv("W2_Superstore_Sales.csv", index_col=0)
superstore
```

```python
superstore.info()
```

```python
superstore['Order_Date'] = pd.to_datetime(superstore['Order_Date'], errors='coerce')
superstore.info()
```

```python
superstore['Ship_Date'] = pd.to_datetime(superstore['Ship_Date'], errors='coerce')
superstore.info()
```

```python
superstore['Order_Year'] = superstore['Order_Date'].dt.year
superstore.info()
```

```python
superstore
```

```python
superstore.head()
```

```python
#Slide 27 Practice
Sales20152018 = superstore[superstore['Order_Year'].between(2015,2018)]['Sales'].sum()
print(f"Sales between 2015 - 2018 is ${Sales20152018:,.2f}.")
```

```python
#Sales by years
Sales_by_years = superstore.groupby('Order_Year')['Sales'].sum()
Sales_by_years.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], edgecolor='black')
plt.title('Total Sales by Year (20152018)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Sales ($)', fontsize=14)
plt.tight_layout()
plt.yticks([100000, 200000, 300000, 400000, 500000, 600000, 700000], ['100k', '200k', '300k', '400k', '500k', '600k', '700k'])

plt.show()
```

```python
orders_by_category = superstore['Category'].value_counts()
orders_by_category.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'], edgecolor='black')
plt.title('Number of Orders by Product Category (20152018)', fontsize=16)
plt.ylabel('Number of Orders', fontsize=14)
plt.xlabel('Product Category', fontsize=14)

plt.show()

#Answer: Office Supplies has the highest amount of orders. 
```

```python
Sales_by_category = superstore.groupby('Category')['Sales'].sum().sort_values(ascending=True) 
Sales_by_category.plot(kind='bar', color=['#ff9999', '#66b3ff', '#99ff99'],edgecolor='black')                      
plt.title('Total Sales by Product Category (20152018)', fontsize=16)
plt.xlabel('Total Sales ($)', fontsize=14)
plt.ylabel('Product Category', fontsize=14)
plt.yticks([100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000], ['100k', '200k', '300k', '400k', '500k', '600k', '700k', '800k'])
plt.tight_layout()
plt.show()

#Answer: Technology has the highest sales volume in USD. 
```

```python
subcategory_sales = superstore.groupby('Sub_Category')['Sales'].sum().sort_values(ascending = True)
subcategory_sales.plot(kind='barh', color='#1f77b4', edgecolor='black')
plt.title('Total Sales by Sub-Category (20152018)', fontsize=16, fontweight='bold')
plt.xlabel('Total Sales ($)', fontsize=14)
plt.ylabel('Sub-Category', fontsize=14)
plt.xticks([50000, 100000, 150000, 200000, 250000, 300000], ['50k', '100k', '150k', '200k', '250k', '300k'])
plt.show()
```

```python
Region_sales = superstore.groupby('Region')['Sales'].sum().sort_values(ascending = False)
Region_sales.plot(kind='bar', color=['#1f77b4'], edgecolor='black')
plt.title('Total Sales by Sub-Category (20152018)', fontsize=16, fontweight='bold')
plt.xlabel('Total Sales ($)', fontsize=14)
plt.ylabel('Sub-Category', fontsize=14)
plt.yticks([100000, 200000, 300000, 400000, 500000, 600000, 700000], ['100k', '200k', '300k', '400k', '500k', '600k', '700k'])
plt.show()
```
