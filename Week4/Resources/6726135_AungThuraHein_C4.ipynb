```python
#Page - 40
#Which countrys GDP is growing during the Year 2010 and Year 2018? Support your answer with visualizations.

gdp = pd.read_csv('WorldBank_GDP.csv')
gdp_2010 = gdp[gdp['Year'] == 2010].set_index('Country Name')['GDP']
gdp_2018 = gdp[gdp['Year'] == 2018].set_index('Country Name')['GDP']

growth = gdp_2018 > gdp_2010
print("Countries with GDP growth from 2010 to 2018:")
print(growth[growth].index.tolist())
countries = ['China', 'Germany', 'Japan', 'United States']
for country in countries:
    country_data = gdp[gdp['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GDP'], marker='o', label=country)

plt.title('GDP Growth (2010-2018)')
plt.ylabel('GDP (USD)')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Find out which country has the highest average temperature

temps = pd.read_csv('temperatures.csv')
country_avg_temps = temps.groupby('country')['avg_temp_c'].max().sort_values(ascending=False)
highest_country = country_avg_temps.index[0]
highest_temp = country_avg_temps.iloc[0]

print(f"Country with highest average temperature: {highest_country} ({highest_temp:.2f}C)")
print(country_avg_temps.head())


#or

print(temps['avg_temp_c'].sort_values(ascending=False))
```

```python
# Find out how many countries where the average temperature is in the range of 20 and 30 Celsius. Show all the countries and their average temperature.

warm_countries = country_avg_temps[(country_avg_temps >= 20) & (country_avg_temps <= 30)]
num_countries = len(warm_countries)
print(f"Number of countries with average temp 20-30C: {num_countries}")
print(warm_countries)
```

```python
#Show the average temperature of Thailand during 2005-01-01 and 2010-01-01 and also find the average temperature during that period.

if 'Unnamed: 0' in temps.columns:
    temps = temps.drop(columns=['Unnamed: 0'])
thai_temps = temps[(temps['country'] == 'Thailand') & (temps['date'] >= '2005-01-01') & (temps['date'] <= '2010-01-01')]
thai_avg = thai_temps['avg_temp_c'].mean()
print(f"Average temperature of Thailand (2005-2010): {thai_avg:.2f}C")
print(thai_temps)
```
