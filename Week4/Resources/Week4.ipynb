# Basic Summary Statistics with Pandas

### Summarizing numberic data
   .mean() .median() .mode() .min() .max() .var() .std() .sum()

```python
print('------------------------ Mean ------')
print(dogs['Height (cm)'].mean())
print('------------------------ Median ----')
print(dogs['Height (cm)'].median())
print('------------------------ Mode ------')
print(dogs['Height (cm)'].mode())
print('------------------------ Min -------')
print(dogs['Height (cm)'].min())
print('------------------------ Max -------')
print(dogs['Height (cm)'].max())
print('------------------------ Variant ---')
print(dogs['Height (cm)'].var())
print('------------------------ Std -------')
print(dogs['Height (cm)'].std())
print('------------------------ Sum -------')
print(dogs['Height (cm)'].sum())
print('------------------------------------')
print(dogs['Height (cm)'].count())
print('------------------------------------')
print(dogs['Height (cm)'].describe())
```

### Sumarizing date
oldest dog:

```python
dogs['Date of Birth'].min()
```

Youngest dog

```python
dogs['Date of Birth'].max()
```

### The .agg() method

```python
def pct30(column) :
    return column.quantile(0.3)
dogs['Weight (kg)'].agg(pct30)
```

### Summaries on multiple Columns

```python
dogs[['Weight (kg)','Height (cm)']].agg(pct30)
```

### Multiple Summaries

```python
def pct40(column):
    return column.quantile(0.4)
dogs['Weight (kg)'].agg([pct30,pct40])
```

### cumulative Sum

```python
print(dogs['Weight (kg)'])
print(dogs['Weight (kg)'].cumsum())
```

```python
# Calc total weekly sales
sales_all =sales["weekly_sales"].sum()
print(sales_all)
# Subset for type A stores, calc total weekly sales
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()
# Subset for type B stores, calc total weekly sales
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()
# Subset for type C stores, calc total weekly sales
sales_C = sales[sales["type"] == "C"]["weekly_sales"].sum()
# Get proportion for each type
print(  [sales_A, sales_B, sales_C]  )
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_propn_by_type)

sales.groupby("type")["weekly_sales"].sum()/ sales_all
```

# Layering Plots

Let’s see what we can do if we have overlapping plots.<br>
Load avoplotto.pkl (pickle file) to avocados dataframe and try the following.<br>

```python
avocados = pd.read_pickle('avoplotto.pkl')
```

```python
# Import matplotlib.pyplot with alias plt
import matplotlib.pyplot as plt

# Look at the first few rows of data
print(avocados.head())

# Get the total number of avocados sold of each size
nb_sold_by_size = avocados.groupby('size')['nb_sold'].sum()

# Create a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind = 'bar')

# Show the plot
plt.show()
```

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby('date')['nb_sold'].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind='line', rot = 45)

# Show the plot
plt.show()
```

```python
# Scatter plot of avg_price vs. nb_sold with title

avocados.plot(x='nb_sold', y='avg_price', kind = 'scatter', title = 'Number of avocados sold vs. average price')

# Show the plot
plt.show()
```

```python
# Histogram of conventional avg_price 
avocados[avocados['type'] == 'conventional']['avg_price'].plot(kind = 'hist')

# Histogram of organic avg_price
avocados[avocados['type'] == 'organic']['avg_price'].plot(kind = 'hist')

# Add a legend
plt.legend(['conventional','organic'])

# Show the plot
plt.show()
```

You can see a complete overlapped plots. The organic blocks the conventional plot.
Try add alpha = 0.6 in plot() and run again.


```python
# Histogram of conventional avg_price 
avocados[avocados['type'] == 'conventional']['avg_price'].plot(kind = 'hist')

# Histogram of organic avg_price
avocados[avocados['type'] == 'organic']['avg_price'].plot(kind = 'hist', alpha=0.6)

# Add a legend
plt.legend(['conventional','organic'])

# Show the plot
plt.show()
```

# Missing Values in DataFrame
- In a pandas DataFrame, missing values are indicated with N-a-N, which stands for "not a number.“


```python
dogs = pd.read_csv('dogs_missing.csv',index_col=0)
dogs
```

### Detecting missing value

```python
dogs.isna()
```

If we chain .isna() with .any(), we get one value for each variable that tells us if there are any missing values in that column.


### Detecting any missing values

```python
dogs.isna().any()
```

### Counting missing values

```python
dogs.isna().sum()
```

### Ploting  missing value

```python
import matplotlib.pyplot as plt
dogs.isna().sum().plot(kind='bar')
plt.show()
```

One option is to remove the rows in the DataFrame that contain missing values. This can be done using the .dropna() method. However, this may not be ideal if you have a lot of missing data, since that means losing a lot of observations.


### Removing missing value

```python
dogs.dropna()
```

### Replacing missing value

```python
dogs.fillna(0)
```
