```python
jul = pd.read_csv('inv_jul.csv')
aug = pd.read_csv('inv_aug.csv')
sep = pd.read_csv('inv_sep.csv')
```

```python
concat_months = pd.concat([jul, aug, sep], keys = ['7Jul', '8Aug', '9Sep'])
print(concat_months)
```

```python
avg_inv_by_month = concat_months.groupby(level=0)['total'].agg('mean')
print(avg_inv_by_month)
```

```python
avg_inv_by_month.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Average Invoice Total by Month')
plt.xlabel('Month')
plt.ylabel('Average Total ($)')
plt.xticks(rotation=0)
plt.show()
```

```python
unemployment = pd.read_csv('unemployment.csv')
inflation = pd.read_csv('inflation.csv')
```

```python
unemployment = unemployment.sort_values('date')
inflation = inflation.sort_values('date')

inflation_unemploy = pd.merge_ordered(inflation, unemployment, on='date', how='inner')

inflation_unemploy = inflation_unemploy[['date', 'cpi', 'unemployment_rate']]
print(inflation_unemploy)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(inflation_unemploy['unemployment_rate'], inflation_unemploy['cpi'], color='blue')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('CPI (Seasonally Adjusted Index)')
plt.title('CPI vs Unemployment Rate (January & June, 20142018)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
corr_matrix = inflation_unemploy[['cpi', 'unemployment_rate']].corr()
print(corr_matrix)
```
